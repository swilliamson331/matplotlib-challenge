### Analysis

- Capomulin and Ramacine were much more effective treatments, in terms of making the tumor smaller, than Infubinol and Ceftamine

- Mouse Weight correlates strongly with tumor volume, as bigger mice have bigger tumors
 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
df = pd.merge(mouse_metadata, study_results, how='right')

# Display the data table for preview
df.head()
```

```python
# Checking the number of mice.
df['Mouse ID'].nunique()
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
dupes = df[df.duplicated(subset=['Mouse ID', 'Timepoint'])]
mouse_dupe = dupes['Mouse ID'].unique()
mouse_dupe
```

```python
# Optional: Get all the data for the duplicate mouse ID.
dupe_df = df.loc[df['Mouse ID'] == "g989"]
dupe_df 
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
dupe_mouse = df[df['Mouse ID'] == 'g989']
dupe_rows = dupe_mouse.index
clean_df = df.drop(dupe_rows, axis=0)
clean_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
clean_df['Mouse ID'].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
# mean, median, variance, standard deviation, and SEM of the tumor volume.
# Assemble the resulting series into a single summary DataFrame.
mean = clean_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
median = clean_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
var = clean_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
stdev = clean_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
count = clean_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].count()
sem = stdev / count**0.5

sum_stats = pd.DataFrame({'Mean': mean,
             'Median': mean,
             'Variance': var,
             'Standard Deviation': stdev,
             'SEM': sem})

sum_stats
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
drug_df = clean_df.groupby('Drug Regimen')
summary_stats = drug_df['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])
summary_stats
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
max_timepoints = clean_df.groupby('Mouse ID')['Timepoint'].max().reset_index(name='MaxTimepoint')


# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merged_df = pd.merge(clean_df, max_timepoints)
final_tumor = merged_df[merged_df['Timepoint'] == merged_df['MaxTimepoint']]
final_tumor
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol = []

# Calculate the IQR and quantitatively determine if there are any potential outliers.

for treatment in treatments:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    tumor_volumes = final_tumor.loc[final_tumor['Drug Regimen'] == treatment, 'Tumor Volume (mm3)']

    # Calculate quartiles, IQR, and bounds for the current treatment
    quartiles = tumor_volumes.quantile([.25, .5, .75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq - lowerq

    lower_bound = lowerq - (1.5 * iqr)
    upper_bound = upperq + (1.5 * iqr) 

    # Calculate outliers using the IQR method
    outliers = tumor_volumes[(tumor_volumes < lower_bound) | (tumor_volumes > upper_bound)]

    # Append the tumor volumes and outlier information to the lists
    tumor_vol.append(tumor_volumes)
    print(f"{treatment}'s potential outliers: {outliers}")
```

```python
# Generate a box plot that shows the distribution of the tumor volume for each treatment group.
plt.boxplot(tumor_vol, flierprops=dict(markerfacecolor='r'))
plt.ylabel('Final Tumor Volume (mm3)')
plt.xticks(range(1, len(treatments) + 1), treatments)
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
capomulin_df = clean_df[clean_df['Drug Regimen'] == 'Capomulin']
mouseI = capomulin_df[capomulin_df['Mouse ID'] == 'b128']

#capomulin_df.head()  (to find a mouse to pick)

plt.plot(mouseI['Timepoint'], mouseI['Tumor Volume (mm3)'])
plt.title('Capomulin treatment of mouse B128')
plt.xlabel('Timepoint (days)')
plt.ylabel('Tumor Volume (mm3)')
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
avg_tumor = capomulin_df.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()
tumor_capo_df = pd.merge(capomulin_df, avg_tumor.reset_index(name='Average Tumor Volume (mm3)'), on='Mouse ID')

plt.scatter(tumor_capo_df['Weight (g)'], tumor_capo_df['Average Tumor Volume (mm3)'])
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
mouse_weight = tumor_capo_df['Weight (g)']
tumor_avg = tumor_capo_df['Average Tumor Volume (mm3)']
correlation = st.pearsonr(tumor_avg,mouse_weight)
pe_slope, pe_int, pe_r, pe_p, pe_std_err = st.linregress(mouse_weight, tumor_avg)
pe_fit = pe_slope * mouse_weight + pe_int

plt.scatter(mouse_weight, tumor_avg)
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.plot(mouse_weight,pe_fit,"-",color="red")
print(f'The correlation between mouse weight and the average tumor volume is {round(correlation[0],2)}')
plt.show()
```
